```python
import pandas as pd
```

# Reading the Datasets

```python
path = '../../data/'
train_dataset = path + 'train.csv'
label_col = 'SalePrice'
```

```python
df_master = pd.read_csv(train_dataset)
df = df_master.copy()
df.head()
```

### Selecting Features

```python
features = ['Foundation', 'KitchenQual', 'TotRmsAbvGrd', 'WoodDeckSF', 'YrSold', '1stFlrSF']
df = df[features + [label_col]]
df.head()
```

### Impute missing values

```python
df.info()
```

No missing values

### Check duplicates

```python
df[features].duplicated(keep='first').sum()
```

- Remove all the duplicates except one

```python
df = df[~df[features].duplicated(keep='first')]
df.tail()
```

```python
df = df.reset_index(drop=True)
df.tail()
```

```python
df_tmp = df_master[features+[label_col]]
df_tmp[df_tmp.duplicated(keep=False)]
```

```python
df_tmp = df_master[features+[label_col]]
df_tmp[df_tmp[features].duplicated(keep=False)]
```

```python
df_tmp2 = df_tmp[df_tmp[features].duplicated(keep=False)]
df_tmp2[(df_tmp2['Foundation'] == 'PConc') & (df_tmp2['KitchenQual'] == 'Gd') & (df_tmp2['TotRmsAbvGrd'] == 6)]
```

### Scale continuous feautres

```python
continuous_columns = df[features].select_dtypes(include='number').columns
continuous_columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[continuous_columns])
```

```python
scaled_columns = scaler.transform(df[continuous_columns])
scaled_columns
```

```python
continuous_features_df = pd.DataFrame(data=scaled_columns, columns=continuous_columns)
continuous_features_df.head()
```

### Categorical features

```python
categorical_columns = df[features].select_dtypes(include='object').columns
categorical_columns
```

```python
categorical_features_df = pd.get_dummies(df[categorical_columns])
categorical_features_df.head()
```

- Join both contineous and categorical dataframes with lable 

```python
final_df = continuous_features_df.join(categorical_features_df).join(df[label_col])
final_df.head()
```

# Model training

- Split dataset

```python
X, y = final_df.drop(columns=[label_col]), final_df[label_col]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

- Train model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

# Model evaluation

```python
y_pred = model.predict(X_test)
```

```python
y_pred[y_pred < 0] = 0
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error

def compute_rmsle(y_test: np.ndarray, y_pred: np.ndarray, precision: int = 2) -> float:
    rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
    return round(rmsle, precision)
```

```python
compute_rmsle(y_test, y_pred)
```
